Defining 30-50 most popular bike stations


```python
df_bicing3['last_reported'] = pd.to_datetime(df_bicing3['last_reported'], unit = 's')
```

```python
df_bicing3.sort_values(['station_id', 'last_reported'], inplace=True)
```

```python
df_bicing3['bike_diff'] = df_bicing3.groupby('station_id')['num_bikes_available'].diff()
```

```python
df_bicing3['checkouts'] = df_bicing3['bike_diff'].apply(lambda x: abs(x) if x < 0 else 0)
checkouts_by_station = df_bicing3.groupby('station_id')['checkouts'].sum().reset_index()
```

```python
# Calcular la tasa de ocupación para cada registro
df_bicing3['occupancy'] = df_bicing3['num_bikes_available'] / (df_bicing3['num_bikes_available'] + df_bicing3['num_docks_available'])

# Calcular la ocupación promedio por estación
occupancy_by_station = df_bicing3.groupby('station_id')['occupancy'].mean().reset_index()

# Combinar ambos indicadores
station_stats = checkouts_by_station.merge(occupancy_by_station, on='station_id')
```

```python
# Ordenar estaciones por checkouts y mostrar las top 30-50
top_checkouts = station_stats.sort_values(by='checkouts', ascending=False).head(50)
print("Top 50 por checkouts:")
print(top_checkouts)

```

```python
# Ordenar estaciones por ocupación (por ejemplo, aquellas que siempre están casi llenas o vacías)
top_occupancy = station_stats.sort_values(by='occupancy', ascending=False).head(50)
print("Top 50 por ocupación:")
print(top_occupancy)
```

```python
desired_num_stations = 50

final_stations = common_top_stations.copy()

for station in top_checkouts['station_id']:
    if len(final_stations) >= desired_num_stations:
        break
    if station not in final_stations:
        final_stations.append(station)

print("Final selected stations:", final_stations)
```

Combining the code into a function

```python
def select_top_stations(df, desired_num_stations=50):
    """
    Filters the input DataFrame to only include the top 'desired_num_stations' stations,
    based on a combination of total checkouts and occupancy rate.

    Parameters:
        df (pd.DataFrame): The input DataFrame containing Bicing data with columns:
                           'last_reported' (unix timestamp), 'station_id', 
                           'num_bikes_available', and 'num_docks_available'.
        desired_num_stations (int): The number of top stations to select.

    Returns:
        final_stations (list): A list of station_ids for the selected top stations.
    """
    # Create a copy to avoid modifying the original DataFrame
    df_copy = df.copy()
    
    # Convert 'last_reported' to datetime
    df_copy['last_reported'] = pd.to_datetime(df_copy['last_reported'], unit='s')
    
    # Sort the DataFrame by station and time
    df_copy.sort_values(['station_id', 'last_reported'], inplace=True)
    
    # Compute the difference in available bikes per station over time
    df_copy['bike_diff'] = df_copy.groupby('station_id')['num_bikes_available'].diff()
    
    # Calculate checkouts: if bikes_available drops, that's a checkout
    df_copy['checkouts'] = df_copy['bike_diff'].apply(lambda x: abs(x) if x < 0 else 0)
    
    # Aggregate total checkouts per station
    checkouts_by_station = df_copy.groupby('station_id')['checkouts'].sum().reset_index()
    
    # Calculate occupancy for each record (occupancy = available bikes / total capacity)
    df_copy['occupancy'] = df_copy['num_bikes_available'] / (
        df_copy['num_bikes_available'] + df_copy['num_docks_available']
    )
    
    # Compute the average occupancy per station
    occupancy_by_station = df_copy.groupby('station_id')['occupancy'].mean().reset_index()
    
    # Merge checkouts and occupancy into a single DataFrame for ranking
    station_stats = checkouts_by_station.merge(occupancy_by_station, on='station_id')
    
    # Rank stations by checkouts
    top_checkouts = station_stats.sort_values(by='checkouts', ascending=False).head(desired_num_stations)
    print(f"Top {desired_num_stations} by checkouts:")
    print(top_checkouts)
    
    # Rank stations by occupancy
    top_occupancy = station_stats.sort_values(by='occupancy', ascending=False).head(desired_num_stations)
    print(f"Top {desired_num_stations} by occupancy:")
    print(top_occupancy)
    
    # Identify common top stations from both rankings
    top_checkout_ids = set(top_checkouts['station_id'])
    top_occupancy_ids = set(top_occupancy['station_id'])
    common_top_stations = list(top_checkout_ids.intersection(top_occupancy_ids))
    print("Common top stations:", common_top_stations)
    
    # Fill up the final station list until the desired count is reached,
    # prioritizing stations from the top checkouts ranking
    final_stations = common_top_stations.copy()
    for station in top_checkouts['station_id']:
        if len(final_stations) >= desired_num_stations:
            break
        if station not in final_stations:
            final_stations.append(station)
    
    print("Final selected stations:", final_stations)
    
    return final_stations

```

Concancating six datasets into one (6 datasets represents 6 months of Bicing data from August 2024 to January 2025)

```python
## Import 6 months data
df_august = pd.read_csv('data/2024_08_Agost_BicingNou_ESTACIONS.csv')
df_september = pd.read_csv('data/2024_09_Setembre_BicingNou_ESTACIONS.csv')
df_october = pd.read_csv('data/2024_10_Octubre_BicingNou_ESTACIONS.csv')
df_november = pd.read_csv('data/2024_11_Novembre_BicingNou_ESTACIONS.csv')
df_december = pd.read_csv('data/2024_12_Desembre_BicingNou_ESTACIONS.csv')
df_january = pd.read_csv('data/2025_01_Gener_BicingNou_ESTACIONS.csv')
```

Adding Weather Data into the dataset

```python
from datetime import datetime
from meteostat import Point, Hourly
import pandas as pd


start = datetime(2024, 8, 1)
end = datetime(2025, 1, 31, 23, 59)


barcelona = Point(41.3851, 2.1734)

# Fetch hourly data for Barcelona
data = Hourly(barcelona, start, end)
data = data.fetch()

# Reset index to convert time to column
data.reset_index(inplace=True)

# Simplify weather conditions using Meteostat's weather codes
# Meteostat weather codes reference: https://dev.meteostat.net/formats.html#weather-condition-codes
weather_mapping = {
    1: 'clear',     # Clear
    2: 'clear',     # Fair
    3: 'cloudy',    # Cloudy
    4: 'cloudy',    # Overcast
    5: 'cloudy',    # Fog
    6: 'rain',      # Freezing Fog
    7: 'rain',      # Light Rain
    8: 'rain',      # Rain
    9: 'rain',      # Heavy Rain
    10: 'rain',     # Freezing Rain
    11: 'rain',     # Heavy Freezing Rain
    12: 'rain',     # Sleet
    13: 'snow',     # Heavy Sleet
    14: 'snow',     # Light Snowfall
    15: 'snow',     # Snowfall
    16: 'snow',     # Heavy Snowfall
    17: 'storm',    # Rain Shower
    18: 'storm',    # Heavy Rain Shower
    19: 'storm',    # Sleet Shower
    20: 'storm',    # Heavy Sleet Shower
    21: 'snow',     # Snow Shower
    22: 'snow',     # Heavy Snow Shower
    23: 'storm',    # Lightning
    24: 'storm',    # Hail
    25: 'storm',    # Thunderstorm
    26: 'storm',    # Heavy Thunderstorm
    27: 'storm'     # Storm
}

# Apply mapping to Meteostat weather codes
data['weather_category'] = data['coco'].map(weather_mapping)

# Keep and rename useful columns
data_final = data[['time', 'temp', 'prcp', 'wspd', 'weather_category']].copy()
data_final.columns = ['timestamp', 'temperature', 'precipitation', 'wind_speed', 'weather_category']

# Save to CSV
data_final.to_csv("barcelona_weather_2024_08_to_2025_01.csv", index=False)

print("Weather data collected and saved.")
print(data_final.head())
```
